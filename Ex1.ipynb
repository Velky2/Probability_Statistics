O professor realizou um censo na turma com algumas caracter√≠sticas relevantes. Nesse caderno, fa√ßo a an√°lise gr√°fica e discuss√£o de tr√™s desses atributos.

```python
#Primeiramente, importemos as bibliotecas e criaremos o dataframe:
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
import seaborn as sns

df = pd.read_excel("Censo Ilum.xlsx")
```

### üçï Gr√°fico de Setores para √°rea com maior interesse

```python
#Primeiro obteremos os dados op√ß√µes escolhidas e quantas vezes foram escolhidas
values = df['Qual destas linhas de estudo √© a que voc√™ tem mais afinidade?'].value_counts()

#Seguidamente, separamos as op√ß√µes (labels) da quantidade (sizes) e armazenamos em listas
labels = values.index.tolist()
sizes = values.values.tolist()

#Agora fa√ßamos o gr√°fico
colors = ['#f58027', '#63ff05', '#27ebf5'] #Cores usadas (hex)
explode = (0.07, 0, 0) #D√° destaque ao maior
plt.pie(sizes, labels=labels, colors=colors, explode=explode, 
        autopct='%1.1f%%', shadow = True, startangle=90) 
#Shadow cria efeito de sombra e startangle corrige angula√ß√£o
plt.title("√Årea de maior interesse na T25 da Ilum") #T√≠tulo
plt.axis('equal') #Garante ser c√≠rculo
plt.show #Mostrar gr√°fico
```

#### An√°lise do Gr√°fico
    √â percept√≠vel pelo gr√°fico de setores a prefer√™ncia dos estudantes pertencentes √† T25 da Ilum pela Ci√™ncia da mat√©ria, com 67.6% de ocorr√™ncia. Por outro lado, menos de 10% dos alunos tem prefer√™ncia por Ci√™ncias Humanas. Essa grande homogeneidade da turma parece colidir com a proposta multidisciplinar da Ilum, mostrando que embora a Ilum se proponha a ter enfoque nas mais diversas √°reas, os ingressantes preferem as exatas. Isso permite duas vias principais de discuss√£o, quanto a se o processo seletivo previlegiou o conhecimento dos alunos nessa √°rea em detrimento das outras ou se a Ilum cria uma imagem √†queles que n√£o a frequentam de maior foco na Ci√™ncia da Mat√©ria.

### ü´í Gr√°fico de Barras sobre Azeitonas

```python
# Primeiro √© necess√°rio armazenar os novos labels e sizes
values = df['Qual a sua opini√£o sobre azeitona?'].value_counts()
labels = values.index.tolist()
sizes = values.values.tolist()

plt.bar(labels, sizes) # Selecionar o que usar para cada eixo
plt.xlabel('Notas') #Legenda do eixo x
plt.ylabel('Frequ√™ncia') #Legenda do eixo y
plt.title('Notas para Azeitona') #T√≠tulo do gr√°fico

plt.show()
grades = []
for i in range(len(sizes)):
    grades += [int(labels[i])] * int(sizes[i]) #Criar lista com as notas com a repeti√ß√£o
print(f'As notas tem m√©dia {st.mean(grades)} e mediana {st.median(grades)}', end = ' ')
print(f'com desvio padr√£o de {st.pstdev(grades)}.')
```

#### As pizzas que o professor comprar√° para a turma final do semestre devem ter azeitona?!
Percebe-se pelo gr√°fico que os valores mais presentes s√£o os extremos, 1 e 5, por√©m, com maior tend√™ncia da turma a gostar, pois, al√©m de 5 ter mais votos que 1, o 4 tem votos significativos. Ao analisar a mediana, percebe-se ser 4, por√©m com alt√≠ssimo desvio padr√£o, 1.7. Assim, caso as pizzas sejam compradas todas com azeitona, a maioria da turma gostaria, por√©m o professor pode criar amargura em uma parcela tamb√©m significativa da turma. Logo, dever√£o ser compradas pizzas com azeitona e sem, para agradar a turma como um todo ou, visando explorar a amizade entre a turma, os que n√£o gostam de azeitona podem ser incentivados a tirar elas de suas fatias e colocarem na de algu√©m que gosta, deixando a todos felizes.

### üìè Gr√°fico de quartis da altura

```python
#Primeiro devemos criar os novos labels e sizes
values = df['Qual √© a sua altura em cent√≠metros?'].value_counts()
labels = values.index.tolist()
sizes = values.values.tolist()
labelsn, sizesn = [], []
for i in range(len(labels)):
    if int(labels[i]) > 3:#Filtro para aqueles com menos de 3cm, filtrando aqueles que confundiram com metro
        labelsn.append(labels[i])
        sizesn.append(sizes[i])

heights = []
for i in range(len(sizesn)):
    heights += [int(labelsn[i])] * int(sizesn[i]) #Criar lista com alturas com repeti√ß√£o

eixo = sns.boxplot(y=heights) #Par√¢metro a ser usado e cria√ß√£o do gr√°fico
eixo.set_ylabel("Altura"); #Legenda do eixo y

print(f'As notas tem m√©dia {st.mean(labelsn)} e mediana {st.median(labelsn)}', end = ' ')
print(f'com desvio padr√£o de {st.pstdev(labelsn)}.')
quartis = st.quantiles(labelsn, n=4)
print(f'Primeiro quartil {quartis[0]} e terceiro quartil {quartis[2]}')
```

### An√°lise do gr√°fico de Alturas
Veja que a mediana encontra-se em 172 cm, com primeiro quartil em 161.5 e terceiro quartil em 178.5. Em mat√©ria do G1 √© apontado que a m√©dia do homem √© 175 e a m√©dia da mulher √© 162. Portanto, √© prov√°vel que os indiv√≠duos sejam em maioria homens, por√©m, como a amostra √© baixa, √© altamente fr√°gil e sucet√≠vel ao acaso.
