# 6. Mister Bin

### 1. Introdução
Faremos uma distribuição binomial com probabilidade p = 0.7 de ganhar e observar sua distribuição ao longo de um número variável n de jogos. 

### 2. pmf

```python
import matplotlib.pyplot as plt
from scipy.stats import binom
p = 0.7
```

O módulo stats.binom.pmf do Scipy faz a pmf automaticamente, usando os parâmetros (k,n,p) com k uma lista de 0 a n, n o número de jogos e p a probabilidade de sucesso e retornando um array com as probabilidades de sucesso em cada k. Nós usaremos esse retorno como a probabilidade do eixo y e no eixo x o array k.

##### 2.1 n = 5

```python
pmf = binom.pmf(range(6), 5, p)

plt.bar(range(6), pmf)
plt.title("pmf para p = 0.7 e n = 5")
plt.xlabel("Número de vitórias")
plt.ylabel("Probabilidade")
plt.show()
```

##### 2.2 n = 10

```python
pmf = binom.pmf(range(11), 10, p)

plt.bar(range(11), pmf)
plt.title("pmf para p = 0.7 e n = 10")
plt.xlabel("Número de vitórias")
plt.ylabel("Probabilidade")
plt.show()
```

##### 2.3 n = 25

```python
pmf = binom.pmf(range(26), 25, p)

plt.bar(range(26), pmf)
plt.title("pmf para p = 0.7 e n = 25")
plt.xlabel("Número de vitórias")
plt.ylabel("Probabilidade")
plt.show()
```

##### 2.4 n = 50

```python
pmf = binom.pmf(range(51), 50, p)

plt.bar(range(51), pmf)
plt.title("pmf para p = 0.7 e n = 50")
plt.xlabel("Número de vitórias")
plt.ylabel("Probabilidade")
plt.show()
```

##### 2.5 n = 100

```python
pmf = binom.pmf(range(101), 100, p)

plt.bar(range(101), pmf)
plt.title("pmf para p = 0.7 e n = 100")
plt.xlabel("Número de vitórias")
plt.ylabel("Probabilidade")
plt.show()
```

### 3. Discussão
Com a plotagem dos gráficos, observamos o esperado, todos os gráficos se apresentaram como uma gaussiana em torno do equivalente a 70% de n. Com o aumento de n, a gaussiana foi ficando mais evidente, pois com números n baixos, como 5, não é possível uma plotagem com gaussiana evidente, porém já apresenta tendência de maior probabilidade em 0.7n e diminuição conforme se afasta disso.

### 4. Conclusão
Logo, obteve-se sucesso em observar gaussianas mais definidas conforme n aumenta, conseguindo-se visualizar como o centro da gaussiana se aproxima sempre de pn.
