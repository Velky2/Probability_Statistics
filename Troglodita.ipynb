# Probablidade e estatística
## Troglodita

São dados dois eventos complementares E1 e E2, ou seja, P(E1) + P(E2) = 1. Além disso tem o evento

1. A probabilidade inicial de E1 é de 20% e X teve acurácia de 99%.

Pelo teorema de bayes:

P(E1|X) = P(X|E1) P(E1)/P(X)

Sabemos que P(E1) = 0.2 e P(X|E1) = 0.99.

Para descobrirmos P(X):

P(X) = P(X|E1)P(E1) + P(X|E2)P(E2) = 0.99 * 0.2 + 0.01 * 0.8 = 0.206

Substituindo em Bayes: P(E1|X) = 0.99 * 0.2/0.206 =~ 0.961

Logo, a probabilidade da explicação 1 após o experimento é de aproximadamente 96.1%.

2. A probabilidade inicial de E1 é de 20% e X teve 50% de acurácia.

Sabemos que P(E1) = 0.2 e P(X|E1) = 0.5.

P(X) = P(X|E1)P(E1) + P(X|E2)P(E2) = 0.5 * 0.2 + 0.5 * 0.8 = 0.5

Com Bayes: P(E1|X) = P(X|E1)P(E1)/P(X) = 0.5 * 0.2/0.5 = 0.2

Logo, a probabilidade da explicação 1 após o experimento é de 20%.

3. A probabilidade inicial de E1 é de 99% e a acurácia de 5%

Sabemos que P(E1) = 0.99 e P(X|E1) = 0.05

P(X) = P(X|E1)P(E1) + P(X|E2)P(E2) = 0.05 * 0.99 + 0.95 * 0.01 = 0.059

Com Bayes: P(E1|X) = P(X|E1)P(E1)/P(X) = 0.05 * 0.99/0.059 =~ 0.839

Logo, a probabilidade da explicação 1 após o experimento é de aproximadamente 83.9%.

#### Façamos agora uma simulação de como se comportaria a probabilidade posteriori da questão 1 após 100 repetições do processo

```python
p = [0.99]
p1 = 0.99
p2 = 1 - p1
x = 0.05
for i in range(9):
    p1 = p1 * x / (p1*x + p2*(1-x))
    p2 = 1 - p1
    p.append(p1)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(10), p)
plt.title("Gráfico de probabilidades inicial e posterioris")
plt.show()
```

Veja como a cada caso ocorre uma grande alteração na probabilidade, começando em 0.99 de probabilidade inicial e na 3° probabilidade posteriori já é de aproximadamente 0%.
