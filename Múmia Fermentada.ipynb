# Múmia Fermentada

### Introdução
O andar do bêbado é o consagrado problema em que um indivíduo dá um passo em uma direção aleatória e acompanhamos seu movimento. Nesse notebook, faremos o andar do bêbado em 2D, ou seja, ele pode se mover para frente, para a direita, para a esquerda ou para trás.

### 1. Simulção e trajetória

Primeiramente, simularei o andar do bêbado e com a biblioteca turtle, acompanharemos toda a trajetória nos primeiros 1000 minutos.

```python
import turtle
from random import randint

# Tela
screen = turtle.Screen()
screen.setup(width=800, height=600)

# Criar uma tartaruga para desenhar os eixos
axis_turtle = turtle.Turtle()
axis_turtle.speed(0)
axis_turtle.color("black")

# Desenhar o eixo X (horizontal)
axis_turtle.penup()
axis_turtle.goto(-400, 0)  # Começa no lado esquerdo do centro
axis_turtle.pendown()
axis_turtle.goto(400, 0)   # Termina no lado direito

# Desenhar o eixo Y (vertical)
axis_turtle.penup()
axis_turtle.goto(0, -300)  # Começa na parte inferior do centro
axis_turtle.pendown()
axis_turtle.goto(0, 300)   # Termina na parte superior

Adicionar marcações nos eixos
step = 50  # Espaçamento entre marcações
for x in range(-400, 401, step):
    axis_turtle.penup()
    axis_turtle.goto(x, -5)
    axis_turtle.pendown()
    axis_turtle.goto(x, 5)

for y in range(-300, 301, step):
    axis_turtle.penup()
    axis_turtle.goto(-5, y)
    axis_turtle.pendown()
    axis_turtle.goto(5, y)

# Opcional: Adicionar rótulos numéricos
axis_turtle.penup()
for x in range(-400, 401, 100):
    axis_turtle.goto(x, -20)
    axis_turtle.write(str(x), align="center")

for y in range(-300, 301, 100):
    if y != 0:  # Evita sobrepor o zero no centro
        axis_turtle.goto(-20, y)
        axis_turtle.write(str(y), align="right")

axis_turtle.hideturtle()  # Esconde a tartaruga após desenhar

# Iniciar simulação
pen = turtle.Turtle()
pen.speed(0)  # Velocidade máxima
pen.pensize(2)
pen.color("blue")

# Posicionar no centro
pen.penup()
pen.goto(0, 0)
pen.pendown()

# Gerar números aleatórios e desenhar o caminho
nums = []
for _ in range(1000):
    nums.append(randint(1, 4))

# Desenhar o caminho baseado nos números aleatórios
for i in nums:
    if i == 1: # Direita
        pen.seth(0)
        pen.forward(5)
    elif i == 2: # Cima
        pen.seth(90)
        pen.forward(5)
    elif i == 3: # Esquerda
        pen.seth(180)
        pen.forward(5)
    else: # Baixo
        pen.seth(270)
        pen.forward(5)
turtle.done()
```

Nesta simulação, geralmente, percebemos a tendência de continuar no mesmo lugar, porém em momentos aleatórios por sorte ir um pouco mais em uma direção. Assim percebemos que, provavelmente, em um grande número de simulações, em média estará na origem após o final.

### 2. Gráfico tempo distância x.

```python
import matplotlib.pyplot as plt
x_pos = []
for i in range(1000):
    x_pos.append(nums[0:i+1].count(1) - nums[0:i+1].count(3)) #Diferença passos direita - esquerda

plt.plot(list(range(1000)), x_pos)
plt.show()
```

Percebemos que olhando em um gráfico de grande intervalo para um pequeno intervalo, geralmente vemos ele estabilizado em um ponto, tendo alguns saltos em momentos específicos

### 3. Gráfico tempo distância y

```python
y_pos = []
for j in range(1000):
    y_pos.append(nums[0:j+1].count(2) - nums[0:j+1].count(4)) # diferença passos cima - baixo

plt.plot(list(range(1000)), y_pos)
plt.show()
```

Novamente percebemos a mesma tendência, intervalos de posição sendo mantida com alguns saltos.

### Coclusão
Portanto, analisando a simulação e os gráficos, percebemos que pela aleatoriedade em um tempo não muito grande, o bêbado consegue se distanciar da origem e terminar alguns passos dela. Porém, caso a simulação fosse feita várias vezes, pelo teorema central do limite, o bêbado estaria na origem. Percebemos que nos gráficos, o bêbado costuma conseguir manter sua posição, mas em alguns momentos consegue se distanciar em uma direção.
